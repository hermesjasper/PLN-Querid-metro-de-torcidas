# Exploracao Inicial - PLN Queridometro de Torcidas

Notebook para carregar a base mock, observar amostras, calcular frequencia de palavras e reservar espaco para TF-IDF e classificacao futura.

```python
from pathlib import Path
from collections import Counter

import matplotlib.pyplot as plt
import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
REAL_DATA_PATH = PROJECT_ROOT / 'data' / 'raw' / 'twitter_public_comments.csv'
MOCK_DATA_PATH = PROJECT_ROOT / 'data' / 'raw' / 'mock_comments.csv'
DATA_PATH = REAL_DATA_PATH if REAL_DATA_PATH.exists() else MOCK_DATA_PATH
DATA_PATH
```

```python
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.info()
df['search_term'].value_counts()
```

```python
STOPWORDS = {
    'o', 'a', 'os', 'as', 'de', 'do', 'da', 'dos', 'das', 'e', 'em', 'um', 'uma',
    'que', 'para', 'por', 'com', 'no', 'na', 'nos', 'nas', 'se', 'esse', 'essa'
}

tokens = []
for text in df['text'].dropna().str.lower():
    tokens.extend(word.strip('.,!?;:()[]') for word in text.split())

word_counts = Counter(word for word in tokens if word and word not in STOPWORDS)
word_counts.most_common(15)
```

```python
top_words = pd.DataFrame(word_counts.most_common(15), columns=['word', 'count'])

ax = top_words.plot(kind='bar', x='word', y='count', legend=False, figsize=(10, 4))
ax.set_title('Frequencia de palavras')
ax.set_xlabel('Palavra')
ax.set_ylabel('Frequencia')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
